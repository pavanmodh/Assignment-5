```python
#Import Libraries
import glob
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

### Visualization function for min max and average

```python
#Visualize Function
def visualize(datasets):
    df = pd.read_csv(datasets)
    temp_df = df
    tempdict = dict()
    # Writing Average, Maximum, Minimum for each feature in a dataset to call later
    for key in temp_df.keys():
        tempdict[key] = list()
        tempdict[key].append(temp_df[key].max())
        tempdict[key].append(temp_df[key].min())
        tempdict[key].append(temp_df[key].mean())      
    final_df = pd.DataFrame.from_dict(tempdict).T
    final_df.columns = ['Max','Min','Average']
    
    #Creating subplots for vizualization for min max and average
    fig,ax = plt.subplots(1,3,figsize = [30,30])
    i = 0
    for col in final_df.keys():
        ax[i].plot(np.array(final_df.index),np.array(final_df[col]))
        ax[i].set_title(col)
        i += 1 
    plt.show(fig) #Plotting the figures
```

### Function to identify errors

```python
#Function to indentiy if there are any issues in the dataset    
def identify_issues(datasets):  
    dataframe = pd.read_csv(datasets)
    if dataframe[dataframe.keys()[0]].max() == 0 and dataframe[dataframe.keys()[20]].max() == 20: #PAssing the fisrt argument
        print('I doubt the integrity of this dataset \n')
    else:
        minimum = np.array([dataframe[key].min() for key in dataframe.keys()]) #Second argument
        minimum_sum = minimum.sum()
        if minimum_sum == 0:
            print('Why this is happening \n')
        else:
            print('Dataset looks good \n')
```

###Loading dataset and calling functions

```python
#Load Datasets
dataset = sorted(glob.glob('data*.csv'))
for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
